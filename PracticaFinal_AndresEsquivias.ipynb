Respecto a las variables Neighbourhood, tenemos 3, que, a priori pro el nombre, van a estar muy correlacionadas; debido a esto, he decidido quedarme con *Neighbourhood Group Cleansed* ya que es más probable que los precios varíen según zonas más amplas, y no tanto por barrios concretos.

````python
import seaborn as sns

# Compute the correlation matrix
corr = np.abs(df_clean.drop(['Price'], axis=1).corr())

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 10))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask,vmin = 0.0, vmax=1.0, center=0.5,
            linewidths=.1, cmap="YlGnBu", cbar_kws={"shrink": .8})

plt.show()
````

Viendo los resultados de la matriz de correlación, habría que eliminar dos de las cuatro variables de disponibilidad ya que tienen una correlación muy alta y son redundantes.

A su vez, la correlación entre camas y persona también es muy alta por lo que hay que eliminar una de ellas. En este caso se va a eliminar *Acommodates*.

````python
df_clean = df_clean.drop(['Accommodates', 'Availability 60', 'Availability 90'], axis = 1)
````

````python
categorical_data = ['Neighbourhood Group Cleansed', 'Property Type', 'Room Type', 'Bed Type']
df_clean = pd.get_dummies(df_clean, columns = categorical_data)
````

````python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_clean, test_size = 0.3, shuffle = True, random_state = 17)

print(f'Dimensiones del dataset de training: {train.shape}')
print(f'Dimensiones del dataset de test: {test.shape}')

train.to_csv('./train.csv', sep=';', decimal='.', index=False)
test.to_csv('./test.csv', sep=';', decimal='.', index=False)

df_train = pd.read_csv('./train.csv', sep=';', decimal='.')
````

Vamos a eliminar de la base de datos aquellas variables que tengan muchos valores Null.
En el caso de 'Cleaning Fee' y 'Security Deposit' parece razonable pensar que si no hay datos es proque esa tasa es 0.

````python
df_train = df_train.drop(['Square Feet', 'Has Availability'], axis = 1)
````

````python
df_train['Security Deposit'] = df_train['Security Deposit'].fillna(0)
df_train['Cleaning Fee'] = df_train['Cleaning Fee'].fillna(0)
````

El resto de variables con algunos NA's los imputaremos con el valor de la media para valores continuos, y la moda para valores discretos.

````python
df_train['Price'] = df_train['Price'].fillna(df_train['Price'].mean())
df_train['Bathrooms'] = df_train['Bathrooms'].fillna(df_train['Bathrooms'].mode()[0])
df_train['Bedrooms'] = df_train['Bedrooms'].fillna(df_train['Bedrooms'].mode()[0])
df_train['Beds'] = df_train['Beds'].fillna(df_train['Beds'].mode()[0])
````

Vamos a proceder a filtrar los datos, ya que es improbable que alguna casa valga 0€ o tenga 0 baños o 0 camas.
En el caso de los dormitorios, puede ser un loft u otro tipo de apartamentos de una sóla habitación, por lo que puede existir esa posibilidad.

````python
bathroom_filter = df_train['Bathrooms'] >= 1
price_filter = df_train['Price'] >= 10
````

````python
df_train = df_train[bathroom_filter]
df_train = df_train[price_filter]
````

##### Test

````python
df_test = pd.read_csv('./test.csv', sep=';', decimal='.')
````

````python
df_test = df_test.drop(['Square Feet', 'Has Availability'], axis = 1)
````

````python
df_test['Security Deposit'] = df_test['Security Deposit'].fillna(0)
df_test['Cleaning Fee'] = df_test['Cleaning Fee'].fillna(0)
df_test['Price'] = df_test['Price'].fillna(df_test['Price'].mean())
df_test['Bathrooms'] = df_test['Bathrooms'].fillna(df_test['Bathrooms'].mode()[0])
df_test['Bedrooms'] = df_test['Bedrooms'].fillna(df_test['Bedrooms'].mode()[0])
df_test['Beds'] = df_test['Beds'].fillna(df_test['Beds'].mode()[0])
````

````python
bathroom_filter = df_test['Bathrooms'] >= 1
price_filter = df_test['Price'] >= 10

df_test = df_test[bathroom_filter]
df_test = df_test[price_filter]
````

##### Modeling

````python
X_train = df_train.drop('Price', axis = 1).values
y_train = df_train['Price'].values
X_test = df_test.drop('Price', axis = 1).values
y_test = df_test['Price'].values
````

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

````python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

n_estimators = range(10, 100, 10)
max_depth = range(1,5)

param_grid = {'n_estimators': n_estimators, 'max_depth': max_depth}

gbr = GradientBoostingRegressor(random_state = 17)
grid = GridSearchCV(gbr, param_grid = param_grid).fit(X_train_scaled, y_train)
````

````python
print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))

# calculamos métricas globales
max_depth_opt = grid.best_params_['max_depth']
n_estimators_opt = grid.best_params_['n_estimators']

error = 1-grid.cv_results_['mean_test_score'].reshape(len(max_depth),len(n_estimators))
colors = ['r','b','g','k','m','c','y', 'w', 'b', 'r']
for i,lr in enumerate(max_depth):    
    plt.plot(n_estimators,error[i,:], colors[i] + '--o', label='md = %g'%lr)

plt.legend()
plt.xlabel('# estimadores')
plt.ylabel('5-fold CV Error')
plt.title('train: %0.3f\ntest: %0.3f'%(grid.score(X_train_scaled,y_train), grid.score(X_test_scaled,y_test)))
plt.grid()
plt.show()
````

Tal y como nos indica el Cross-Validation, y que podemos ver en la gráfica, los mejores hiper parámetros son 100 estimadores y una profundidad de 4.
Sin embargo, hay cierto overfitting, ya que la diferencia entre trainy test es algo grande; por lo que deberíamos hacer una selección de características para ver si podemos reducir la complejidad del modelo y que generalice mejor.
Además, podemos ver que con menos estimadores (50) conseguimos una precisión parecida, por lo que podría ayudarnos a reducir la complejidad del modelo y, por ende, su varianza.

````python
mdOptimo = grid.best_params_['max_depth']
neOptimo = 50
gbr = GradientBoostingRegressor(n_estimators = neOptimo, max_depth = mdOptimo,
                                 random_state = 17).fit(X_train_scaled, y_train)

print("Train: ", gbr.score(X_train_scaled, y_train))
print("Test: ", gbr.score(X_test_scaled, y_test))

importances = gbr.feature_importances_
importances = importances / np.max(importances)
features = df_train.columns.drop('Price')
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10,10))
plt.barh(range(X_train_scaled.shape[1]), importances[indices])
plt.yticks(range(X_train_scaled.shape[1]), features[indices])
plt.show()
````

Como podemos ver en el gráfico, hay una gran cantidad de variables poco importantes; estas variables incluyen, la mayoría de las opciones de 'Property Type'; todas las opciones de 'Bed Type' y algunas de las opciones de Neighbourhood Group Cleansed; esto nos puede indicar que, no hay suficientes datos en algunos barrios para que tenga la suficiente relevancia en el modelo, o, que dependiendo del barrio no hay apenas diferencia; esto es, entre Salamanca y Vallecas si habr´´a una diferencia relevante pero entre Arganzuela y Tetuan no.

Para solucionar esto, se podrían agrupar aún más los barrios en grupos más grandes, que engloben zonas con un mismo nivel económico y así tener más datos por zona.

Se va a elegir las variables más relevantes y reentrenar de nuevo a ver si reducimos la varianza.

````python
X_train_importance = X_train_scaled[:, indices[:13]]
X_test_importance = X_test_scaled[:, indices[:13]]

gbr = GradientBoostingRegressor(n_estimators = neOptimo, max_depth = mdOptimo,
                                 random_state = 17).fit(X_train_importance, y_train)

y_pred = gbr.predict(X_test_importance)

print("Train: ", gbr.score(X_train_importance, y_train))
print("Test: ", gbr.score(X_test_importance, y_test))
print("Test rmse: ", np.sqrt((y_test - y_pred)**2).mean())
````

Aunque, nos disminuye un poco la precisión del modelo; dependiendo del caso de uso podría ser interesante reducir la complejidad de 62 variables a sólo 13

Una vez elegidos todos los hiper parámetros, y las nuevas variables habría que re-entrenar todo el dataset de nuevo juntando los conjuntos de training y test

````python
X = np.concatenate([X_train_importance, X_test_importance])
y = np.concatenate([y_train, y_test])
gbr = GradientBoostingRegressor(n_estimators = 100, max_depth = 4,
                                 random_state = 17).fit(X, y)

print("Score: ", gbr.score(X, y))
````

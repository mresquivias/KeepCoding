````python
# Let's create a list with image url and index
images_paths = []
for i, img_url in enumerate(df_train['Thumbnail Url']):
  images_paths.append([i, img_url])
````

````python
import imageio as io
import cv2

# This function downloads the images and returns them with the index where to save them

def get_image(data_url, target_size = (112, 112)):
    idx, url = data_url
    try:
        img = io.imread(url)
        
        # if channel info is in the first place, move to the last place
        if (img.shape[0] < img.shape[1]) and (img.shape[0] < img.shape[2]):
            img = moveaxis(img, 0, 2)
        
        # In case any channel is RGBA take only RGB 
        img = img[:, :, :3]
        # Some images may be in black & white. I convert to to RGB as otherwise it generates error
        if img.ndim < 3:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        img = cv2.resize(img, dsize = target_size)
        return img, idx
    except IOError as err:
        return (None, idx)
````

### Modeling

````python
from keras import layers
from keras.layers import Input, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D
from keras.layers import AveragePooling2D, MaxPooling2D, Dropout, GlobalMaxPooling2D, GlobalAveragePooling2D
from keras.layers import concatenate
from keras.models import Model
````

````python
# define two sets of inputs
inputA = Input(shape = (62,))
inputB = Input(shape = (112, 112, 3))

# Model for numerical data
X = Dense(32, activation = "relu", name = 'fc0')(inputA)
X = Dropout(0.3, name = 'drop0')(X)
X = Dense(16, activation = "relu", name = 'fc1a')(X)
X = Model(inputs = inputA, outputs = X)

# Model for image data
Y = ZeroPadding2D((3, 3), name = 'padding0')(inputB)
Y = Conv2D(16, 3, name = 'conv0')(Y)
Y = BatchNormalization(axis = 3, name = 'norm0')(Y)
Y = Activation('relu')(Y)
Y = MaxPooling2D(name = 'pool0')(Y)

Y = ZeroPadding2D((2, 2), name = 'padding1')(inputB)
Y = Conv2D(32, 3, name = 'conv1')(Y)
Y = BatchNormalization(axis = 3, name = 'norm1')(Y)
Y = Activation('relu')(Y)
Y = MaxPooling2D(name = 'pool1')(Y)

Y = Flatten(name = "flatten")(Y)
Y = Dense(16, activation = 'relu', name = 'fc1b')(Y)
Y = Model(inputs = inputB, outputs = Y)

# combine the output of the two branches
combined = concatenate([X.output, Y.output])

# apply a FC layer and then a regression prediction on the
Z = Dense(16, activation = "relu", name = 'fc2')(combined)
Z = Dropout(0.3, name = 'drop1')(Z)
Z = Dense(8, activation="relu", name = 'fc3')(Z)
Z = Dense(1, activation = "relu", name = 'fc4')(Z)

model = Model(inputs = [X.input, Y.input], outputs = Z)
````

Este modelo, al aceptar imágene sy adtos numéricos tiene dos partes:
* Una red con varias capas fully-connected que tratan los datos numéricos.
* Una red convolucional para procesar las imágenes y extraer características

Una vez tenemos procesadas las características de ambas redes, concatenamos esas caracterís'meanicas, añadimos unas capas más fully-connected y ponemos como función de pérdida, al querer hacer una regresión, mean squared error.

````python
model.summary()
````

````python
model.compile(loss = 'mean_squared_error', optimizer ='adam', metrics = ["mse"])
````

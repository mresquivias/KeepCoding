### Reporting

El modelo final elegido para el análiss de sentimiento ha sido una Regresión Logística, con el cual hemos obtenido estos resultados:

    Classification report:
             precision    recall  f1-score   support

          0       0.99      0.86      0.92    276976
          1       0.30      0.87      0.45     19233

    avg / total       0.95      0.86      0.89    296209


    
   

A la hora de tratar con este tipo de métricas es muy importante prestar atención al caso de uso y ver qué es más importante; esto es, aunque siempre vamos a querer tener cuantos menos errores (falsos positivos y falsos negativos) posibles, a veces es necesario hacer un tradeoff entre ellos y, según donde pongamos el umbral, obtendremos más de unos o de otros.

En este caso, y teniendo en cuenta que estamos analiazando reviews de Amazon, lo importante para nuestro negocio sería identificar correctamente aquellas reviews negativas e intentar que no se nos pase ninguna para poder recoger ese feedback y mejorar nuestro producto para evitar tener más clientes descontentos en el futuro.

Sumado a esto, podemos observar que tenemos un problema de desbalanceo de clases (276976 reviews positivas frente a 19233 negativas).
Debido a ello, debemos interpretar las métricas con cuidado, ya que, si el modelo simplemente seleccionara todas las reviews como positivas, obtendría una precisión del 93.5%.

Por eso, y sumado al caso de uso que tenemos, es más importante centrarnos en las métricas de las reviews negativas, de las cuale spodemos observar un *recall* de **0.87**, lo que nos indica que, de todas las posibles reviews negativas, hemos identificado correctamente un **87%**

Uaa vez elegido el modelo, como se menciona arriba, debemos seleccionar aquel umbral que nos permita obtener las mejores métricas para el caso de uso que tenemos:
<img src='prec_rec_plot.png'>
Si lleváramos el umbral a 0, nuestro *recall* sería de **1.00**, por lo que identificaríamos correctamente todas las reviews negativas, en detrimento de la precisión, ya que tendríamos muchos más falsos nefativos.

En el caso contrario, con el umbral a 1, obtendríamos una precisión del **1.00** pero el *recall* sería **0**.

En nuestro caso el umbral que consiguiera maximizar mabas métricas sería, aproximandamente, un **0.85**.

Aunque nuestro modelo de análisis de sentimiento ha conseguido unos resultados aceptables, sería importante teer en cuenta una serie de aspectos y pasos futuros para mejorarlo:
* Solucionar el desbalanceo de clases añadiendo más reviews negativas.
* Comprobar si mantiene estos resultados con datasets diferentes y es capaz de generalizar.
* De ser así, utilizar diferentes datasets para ayudar a esa genralización.
